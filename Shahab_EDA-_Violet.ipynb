## 5. Make 5 questions and solve them by using group by.

```python
# Q1. What is the total enrollment for each study status?

Enrollment_by_status = DF_Covid.groupby('Status')['Enrollment'].sum()

print('Total enrollment for each study status:')
print(Enrollment_by_status)
```

```python
# Q2. What is the maximum enrollment for studies in each phase?

Max_enrollment_by_phase = DF_Covid.groupby('Phases')['Enrollment'].max()

print('Maximum enrollment for studies in each phase:')
print(Max_enrollment_by_phase)
```

```python
# Q3. What is the average enrollment for studies by Age?

Average_enrollment_by_age = DF_Covid.groupby('Age')['Enrollment'].mean()

print('Average enrollment for studies by Age:')
print(Average_enrollment_by_age)
```

```python
# Q4. What is the total number of studies for each gender?

Studies_by_gender = DF_Covid.groupby('Gender')['NCT Number'].count()

print('Total number of studies for each gender:')
print(Studies_by_gender)
```

```python
# Q5. What is the total enrollment by gender.

Studies_by_gender = DF_Covid.groupby('Gender')['Enrollment'].sum()

print('Total number of enrollment by gender:')
print(Studies_by_gender)
```

## Make 5 questions and solve them by using conditional statements

```python
# Q1. How many trials are completed?

Completed_trials_count = DF_Covid[DF_Covid['Status'] == 'Completed'].shape[0]

print('Number of trials completed:')
print(Completed_trials_count)
```

```python
# Q2. What is the average enrollment of participants in completed studies?

Average_Enrollment_in_completed_studies = DF_Covid.loc[DF_Covid['Status'] == 'Completed', 'Enrollment'].mean()

print('Average enrollment of participants in completed studies:')
print(Average_Enrollment_in_completed_studies)
```

```python
# Q3. How many studies have enrollment greater than 50?

Studies_with_enrollment_greater_than_50 = DF_Covid[DF_Covid['Enrollment'] > 50].shape[0]

print('Number of studies with enrollment greater than 50:')
print(Studies_with_enrollment_greater_than_50)
```

```python
# Q4. What is the title of the trial with the highest enrollment?

Trial_with_highest_enrollment = DF_Covid.loc[DF_Covid['Enrollment'].idxmax(), 'Title']

print('Title of the trial with the highest enrollment:')
print(Trial_with_highest_enrollment)
```

```python
# Q5. How many unique conditions have trials with enrollment less than 30?

Unique_conditions_with_enrollment_less_than_30 = DF_Covid[DF_Covid['Enrollment'] < 30]['Conditions'].nunique()
print('Number of unique conditions with enrollment less than 30:')
print(Unique_conditions_with_enrollment_less_than_30)
```
